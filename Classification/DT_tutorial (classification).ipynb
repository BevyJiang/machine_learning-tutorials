```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import graphviz
```

```python
wine = load_wine()
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data), pd.DataFrame(wine.target)],axis=1).head()
```

```python
#train test split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size = 0.3)
```

```python
#train a tree model
clf = tree.DecisionTreeClassifier(criterion = 'entropy', random_state=30)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```python
#draw a decision tree graph
feature_name = ['alcohol','malic_acid','ash','alcalinity_of_ash','magnesium','total_phenols','flavanoids','nonflavanoid_phenols',
 'proanthocyanins','color_intensity','hue','od280/od315_of_diluted_wines','proline']
wine_data = tree.export_graphviz(clf, 
                               feature_names = feature_name,
                               class_names = ['0','1','2'],
                               filled = True,
                               rounded = True)
graph = graphviz.Source(wine_data)
graph
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state=30
                                  ,splitter="random")
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```python
wine_data = tree.export_graphviz(clf
                                ,feature_names= feature_name 
                                ,class_names=["0","1","2"] 
                                ,filled=True
                                ,rounded=True
                               )
graph = graphviz.Source(wine_data)
graph
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state=30
                                  ,splitter="random"
                                  ,max_depth=3
                                  ,min_samples_split=10
                                 )

clf = clf.fit(Xtrain, Ytrain)
wine_data = tree.export_graphviz(clf
                               ,feature_names= feature_name
                               ,class_names=["0","1","2"] 
                               ,filled=True
                               ,rounded=True
                               )

graph = graphviz.Source(wine_data)
graph
```

```python
score = clf.score(Xtest, Ytest)
score
```
